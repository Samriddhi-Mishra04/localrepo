```python
# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text
```

```python
# Load and extract text from the PDF
pdf_path = "Indian-Polity-6th-Edition-M-Laxmikant-booksfree.org_.pdf"  # Replace with the actual PDF path
pdf_text = extract_text_from_pdf(pdf_path)
```

```python
# Continuous Question-Answer Loop
print("Ask me anything about the document (type 'quit' to exit):")
while True:
    user_input = input("\nYour question: ").strip()

    if user_input.lower() == "quit":
        print("Exiting... Have a great day!")
        break

    # Send request to Hugging Face API
    data = {"inputs": f"Based on the following document, answer: {user_input}\n\nDocument: {pdf_text[:5000]}"}  # Limit text size
    response = requests.post(API_URL, headers=HEADERS, json=data)

    # Print response
    output = response.json()

    if isinstance(output, dict) and "error" in output:
        print(f"Error: {output['error']}")
    else:
        print("\nAnswer:", output[0]["generated_text"])

```
